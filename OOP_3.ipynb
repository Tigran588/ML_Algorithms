```python
#3#
from abc import ABC,abstractmethod

class Dish(ABC):
  def __init__(self,name :str, price: str):
    self.name = name
    self.price = price

  def get_attr(self):
    return self.name,self.price


class BBQ(Dish):
  def __init__(self,name,price):
    super().__init__(name,price)

class Apitizers(Dish):
  def __init__(self,name,price):
    super().__init__(name,price)

class Menu:
  def __init__(self):
    self.menu = []

  def add_dish(self,dish):
    self.menu.append((dish.name,dish.price))

  def view_menu(self):
    for i in self.menu:
       print(i)

dish1 = BBQ('chicken',1200)
dish2 = Apitizers('kesar',750)

men = Menu()
men.add_dish(dish1)
men.add_dish(dish2)
#men.view_menu()


class Order:
  def __init__(self,customer,menu):
    self.order_price = 0
    self.orders = []
    self.cutomer = customer
    self.menu = menu

  def make_order(self,dish_name:str):
    for item in self.menu.menu:
      if dish_name in item[0]:
        print(f'ka senc uteliq {dish_name}')
        self.orders.append(item[0])
        self.order_price += (item[1])

  def get_order_attr(self):
    return self.orders,self.order_price

ord = Order('Tiko',men)

class Customer:
  def __init__(self,name:str,contact:str)-> None:
    self.contact = contact
    self.name = name

  def View_menu(self,menu:Menu):
    menu.view_menu()

  def customer_order(self,order:Order,dish_name:str):
    order.make_order(dish_name)

  def view_order_history(self,order:Order):
    return order.get_order_attr()

cust = Customer('Tiko','hamar')

#cust.View_menu(men)
cust.customer_order(ord,'kesar')
cust.view_order_history(ord)
```
